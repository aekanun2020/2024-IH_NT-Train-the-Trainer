```python
### https://chat.openai.com/share/c6acacd2-7674-40a7-a3bf-6545e622e020

import pandas as pd

# Assuming `data.csv` is your file name
df = pd.read_csv('../data/LoanStats_web.csv', dtype=str)
```

```python
df.columns
```

```python
# Function to check for columns to exclude
def check_columns(df):
    exclude_cols = []
    for col in df.columns:
        max_consecutive_nans = df[col].isna().astype(int).groupby(df[col].notna().astype(int).cumsum()).sum().max()
        #print(max_consecutive_nans)
        if max_consecutive_nans > 3:
            exclude_cols.append(col)
    return exclude_cols

# Exclude columns with more than 3 consecutive NAs
cols_to_exclude = check_columns(df)
print(cols_to_exclude)
df_filtered = df.drop(columns=cols_to_exclude)

# df_filtered now contains the DataFrame without the specified columns
```
